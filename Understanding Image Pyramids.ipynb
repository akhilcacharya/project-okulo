```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("Lenna_color.png",0)
print img.shape
plt.subplot(1,2,1),plt.imshow(img, cmap="gray"),plt.title("Original"),plt.xticks([]),plt.yticks([])
plt.show()
```

```python
pd1 = cv2.pyrDown(img)
print pd1.shape
plt.subplot(1,2,1),plt.imshow(pd1, cmap="gray"),plt.title("Pyramid Down 1"),plt.xticks([]),plt.yticks([])
plt.show()
```

```python
pd2 = cv2.pyrDown(pd1)
pd2 = cv2.pyrDown(pd2)
pd2 = cv2.pyrDown(pd2)
print pd2.shape
plt.subplot(1,2,1),plt.imshow(pd2, cmap="gray"),plt.title("Pyramid Down n"),plt.xticks([]),plt.yticks([])
plt.show()
```
